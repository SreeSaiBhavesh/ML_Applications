# Churn Prediction

**Churn rate is a critical metric of customer satisfaction. Low churn rates mean happy customers, high churn rate means customers are leaving the company.**
Churn rate tells us how many already existing customers are leaving the product/business. Churn is important factor for business growth and analysis of metrics.

 Churn rate = (No Of Customers lost during a time frame / No Of Customers at the beginning of the time frame) *100 

   The impact of the churn rate is clear, so we need strategies to reduce it. Predicting churn is a good way to create proactive marketing campaigns targeted at the customers that are about to churn

# Data Preprocessing

```python
# Unnecessary column here is CustomerID
data.drop(["customerID"],axis=1,inplace=True)
```

```python
def binary_map(feature):
    return feature.map({'Yes':1, 'No':0})

## Encoding target feature
data['Churn'] = data[['Churn']].apply(binary_map)

# Encoding gender category
data['gender'] = data['gender'].map({'Male':1, 'Female':0})

#Encoding other binary category
binary_list = ['SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
data[binary_list] = data[binary_list].apply(binary_map)

#Encoding the other categoric features with more than two categories
data = pd.get_dummies(data, drop_first=True)
```

# Min Max Scaling of the features

Except **tenure, Total Charges and Monthly charges** all the columns are scaled, So applyinh MinMax Scale for these 3 columns

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
data['tenure'] = sc.fit_transform(data[['tenure']])
data['MonthlyCharges'] = sc.fit_transform(data[['MonthlyCharges']])
data['TotalCharges'] = sc.fit_transform(data[['TotalCharges']])
```

# Splitting the data into train and test sets

```python
from sklearn.model_selection import train_test_split
X = data.drop('Churn',axis=1)
y = data['Churn']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train = X_train.fillna(0)
```

# Performing Different Machine Learning algorithms

# 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
lg=LogisticRegression()
lg.fit(X_train,y_train)
lg_predict=lg.predict(X_test)
print(accuracy_score(y_test,lg_predict))
```

# 2. Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
rfc_predict = rfc.predict(X_test)
print(accuracy_score(y_test,rfc_predict))
```

# 3. Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
dtc_predict = dtc.predict(X_test)
print(accuracy_score(y_test,dtc_predict))
```

# 4. Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
nb_predict = nb.predict(X_test)
print(accuracy_score(y_test,nb_predict))
```

# 5. Support Vector Machine

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train)
predict = svc.predict(X_test)
print(accuracy_score(y_test,predict))
```

# 6. Random Forest using RandomCV

```python
param_grid={'n_estimators':[int(x) for x in np.linspace(start=200,stop=1200,num=11)],
            'max_features':['auto','sqrt'],
            'max_depth':[int(x) for x in np.linspace(start=10,stop=100,num=11)],
            'min_samples_leaf':[1,2,3,5],
            'min_samples_split':[2,5,10,15]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_cv=RandomizedSearchCV(rfc,param_grid,cv=3,verbose=2,random_state=42)
random_cv.fit(X_train,y_train)
best_random=random_cv.best_estimator_
prediction_cv=best_random.predict(X_test)
print(accuracy_score(y_test,prediction_cv))
```
