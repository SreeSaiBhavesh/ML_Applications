```python
# Importing csr_matrix to compress the sparse data
from scipy.sparse import csr_matrix
csr_data = csr_matrix(data.values)
print(csr_data)
```

```python
data.reset_index(inplace=True)
```

# Implementing KNN Algorithm to give nearer results or recommendation of movies for movie search by user

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', n_neighbors=20, n_jobs=-1)
```

```python
knn.fit(csr_data)
```

```python
def movie_recommendation(movie_name):
    n_moviesRecommend = 10
    movie_list = movies[movies['title'].str.contains(movie_name)]  
    if len(movie_list):        
        dum_movie= movie_list.iloc[0]['movieId']
        dum_movie = data[data['movieId'] == dum_movie].index[0]
        distances , indices = knn.kneighbors(csr_data[dum_movie],n_neighbors=n_moviesRecommend+1,return_distance=True)    
        rec_movie_indices = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[:0:-1]
        recommend_frame = []
        for val in rec_movie_indices:
            dum_movie = data.iloc[val[0]]['movieId']
            idx = movies[movies['movieId'] == dum_movie].index
            recommend_frame.append({'Title':movies.iloc[idx]['title'].values[0],'Distance':val[1]})
        res = pd.DataFrame(recommend_frame,index=range(1,n_moviesRecommend+1))
        return res
    else:
        return "No movies found. Please check your input"
```

```python
movie_recommendation('Memento')
```
